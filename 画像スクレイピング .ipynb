```python
import requests
from bs4 import BeautifulSoup
import re
```

### 読み込んだ画像をデコードして表示する
- 例として、Googleのトップページ画像を取得している
- 参考ページ:
    - https://stackoverflow.com/questions/46429088/how-to-view-an-image-with-matplotlib-when-using-requests-getimage-url

```python
# get methodでリンク先のcontentを取得する
img_url = "https://www.google.co.jp/images/branding/googlelogo/2x/googlelogo_color_272x92dp.png"
response = requests.get(img_url)
print("status:", response.status_code)
print(response.content[:100], "...")

from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

# 文字列をデコードして画像データに変換する
i = Image.open(BytesIO(response.content))
plt.imshow(i)
plt.show()
```

### 画像処理用の関数を自作

```python
def extract_hostname(url):
    # 正規表現を使い、画像URLを取得する
    # https://note.nkmk.me/python-re-match-object-span-group/#group
    host_name = re.search(r"https?://(.*?)/", url_name + "/").group()
    return host_name[:-1]

def does_start_with_http(url):
    if re.search(r"https?://", url):
        return True
    return False

def plot_get_image(url):
    response = requests.get(url)
    try:
        i = Image.open(BytesIO(response.content))
        plt.imshow(i)
        plt.show()
    except:
        print("Failed to show image!")
```

## モスバーガーのメニュー写真を取得する

```python
url_name = "https://www.mos.jp/menu/"
host_name = extract_hostname(url_name)

print(host_name)
url = requests.get(url_name)
soup = BeautifulSoup(url.content, "lxml")

elems = soup.find_all("img")

for idx, elem in enumerate(elems): 
        print(elem)
        try:
            elem['src']
        except KeyError:
            continue
            
        if does_start_with_http(elem['src']):
            img_url = elem['src']
        else:
            img_url = host_name + elem['src']
        
        print(img_url)
        plot_get_image(img_url)
```
